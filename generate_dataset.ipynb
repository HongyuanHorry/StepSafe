```python
import pandas as pd
import random
import os
from sklearn.model_selection import train_test_split

# ----------------------------
# SET YOUR FOLDER PATH
# ----------------------------
folder_path = "/content/drive/MyDrive/FIT5120_Datasets/"
os.makedirs(folder_path, exist_ok=True)

# ----------------------------
# CONFIG
# ----------------------------
NUM_ROWS = 50000

job_types = ["Casual", "Part-time", "Full-time", "Remote"]
platforms = ["WhatsApp", "Gmail", "LinkedIn", "Telegram", "SMS"]
message_types = ["SMS", "Email"]

scam_types = ["Task-Based Scam", "Phishing", "Fake Job", "Advance Fee"]

legit_templates = [
    "We received your application. Please attend interview at our Melbourne office.",
    "Your resume has been shortlisted. We will contact you soon.",
    "Please submit your portfolio via our official company website.",
    "Interview scheduled for next week. No payment required.",
    "We would like to invite you for a formal interview.",
    "Thank you for applying. We will review your application and contact you.",
"Your interview is scheduled for next week at our office.",
"Please submit your portfolio through our official website.",
"We are pleased to invite you for a formal interview process.",
"Your application has been successfully received.",
"We will notify you once the review process is complete.",
"Please log in to our official portal to track your application.",
"You can apply through our verified company careers page.",
"Our HR team will reach out if shortlisted.",
"Interview will be conducted via official company email.",
"No fees are required at any stage of recruitment.",
"All communication will be via official domain emails.",
"Please attend the scheduled interview session.",
"We only hire through official recruitment channels.",
"Thank you for your interest in our company.",
"Our recruitment process does not require payments.",
"We will inform you about next steps soon.",
"Kindly wait for confirmation from our HR team.",
"Your application is under review.",
"We appreciate your patience during the hiring process."
]

scam_templates = [
    "Earn ${}/day by completing simple tasks from home.",
    "Pay ${} onboarding fee to secure your job immediately.",
    "Congratulations! You are selected. Share your bank details now.",
    "Limited offer! Earn money instantly. Click here {}",
    "Work from home and earn ${}/week. No experience required.",
    "Earn $450/day reviewing apps from home, no skills needed, start today.",
"Immediate hiring! Earn $200 daily, simple online tasks, apply now.",
"Pay $80 fee to unlock premium job access and start earning instantly.",
"Congratulations, you are shortlisted. Provide bank details for salary setup.",
"Limited slots available! Register now and earn up to $1000 weekly.",
"Earn money by liking posts, daily payout guaranteed.",
"Urgent hiring, no interview required, click link to begin.",
"Send your ID and bank details to confirm your remote job.",
"Work from home and earn fast cash, no experience needed.",
"Quick job offer, pay small fee and start earning immediately.",
"Earn $300 daily typing simple data online.",
"Only 5 positions left, secure your job now.",
"Pay onboarding fee and get instant access to job dashboard.",
"Share your PayPal details to activate your payment account.",
"Earn passive income by completing easy tasks online.",
"Make money watching videos from home.",
"Start earning instantly, no experience required.",
"Get paid daily for reviewing products online.",
"Exclusive opportunity, limited positions, act now.",
"Secure your job today with a small refundable deposit.",
"Earn weekly income with flexible hours.",
"Quick approval job, no background check required.",
"Earn money from home using your phone.",
"Join now and start earning immediately.",
"Fast cash opportunity, limited offer.",
"Simple online work, high income guaranteed.",
"Work anytime, earn daily income.",
"Easy job, no skills required.",
"Get paid instantly after task completion.",
"Remote job available, apply now before closing."
]

links = [
    "http://quicktask-au.com",
    "http://earnfast.click",
    "http://jobfast-au.net",
    "http://taskhub-au.com",
    "http://quickearn-au.net",
"http://taskrewardhub.com",
"http://fastpay-jobs.net",
"http://earnonline-au.click",
"http://jobportal-fast.com",
"http://remotework-au.net",
"http://clicktaskpro.com",
"http://easyincome-au.com",
"http://earncashnow.click",
"http://jobaccess-now.net"
]

# ----------------------------
# GENERATE DATA
# ----------------------------
def generate_row(i):
    is_scam = random.choice([0, 1])

    job_type = random.choice(job_types)
    platform = random.choice(platforms)
    msg_type = random.choice(message_types)

    if is_scam:
        template = random.choice(scam_templates)
        value = random.randint(200, 2000)
        link = random.choice(links)

        if "{}" in template:
            text = template.format(value if "$" in template else link)
        else:
            text = template.format(value)

        return {
            "message_id": i,
            "text": text,
            "message_type": msg_type,
            "platform": platform,
            "job_type": job_type,
            "contains_link": 1 if "http" in text else 0,
            "contains_email": 0,
            "has_payment_request": 1 if "pay" in text.lower() else 0,
            "asks_personal_info": 1 if "bank" in text.lower() else 0,
            "scam_type": random.choice(scam_types),
            "label": 1
        }
    else:
        text = random.choice(legit_templates)

        return {
            "message_id": i,
            "text": text,
            "message_type": msg_type,
            "platform": platform,
            "job_type": job_type,
            "contains_link": 0,
            "contains_email": 0,
            "has_payment_request": 0,
            "asks_personal_info": 0,
            "scam_type": "Legitimate",
            "label": 0
        }

# Create dataset
data = [generate_row(i) for i in range(NUM_ROWS)]
df = pd.DataFrame(data)

# ----------------------------
# SAVE FULL DATASET
# ----------------------------
df.to_csv(folder_path + "scam_messages_full.csv", index=False)

# ----------------------------
# TRAIN / TEST SPLIT
# ----------------------------
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

train_df.to_csv(folder_path + "scam_messages_train.csv", index=False)
test_df.to_csv(folder_path + "scam_messages_test.csv", index=False)

print("All datasets saved to Google Drive!")
```
